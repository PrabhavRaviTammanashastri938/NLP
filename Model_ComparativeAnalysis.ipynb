```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set_palette('rainbow')
sns.set_style('darkgrid')
```

```python
df = pd.read_excel(r'clinical_NER_perfomance_models.xlsx')
df
```

```python
df['Model-Code'] = ['SciBERT','Albert-NER','BERT-Cased','DistillBERT','BioBERT-Cased']
```

```python
df
```

```python
colors = ['blue', 'green', 'orange', 'red','yellow']
```

### Performance Evaluation of Models

```python
plt.figure(figsize=(8,5))
plt.bar(df['Model-Code'],df['Accuracy'],color=colors,edgecolor='black')
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.title("Accuracy of Different Clinical NER Models")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Model-Code'],df['Precision'],color=colors,edgecolor='black')
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.title("Precision of Different Clinical NER Models")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Model-Code'],df['F1 Score'],color=colors,edgecolor='black')
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.title("F1 Score of Different Clinical NER Models")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Model-Code'],df['Recall'],color=colors,edgecolor='black')
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.title("Recall of Different Clinical NER Models")
plt.show()
```

```python
plt.figure(figsize=(10,5))
data1 = df['Accuracy']
data2 = df['Precision']
data3 = df['F1 Score']
data4 = df['Recall']
x_values = range(len(data1))

plt.title("All Classification Metrics of Different Clinical NER Models")
plt.bar(x_values, data1, width=0.1, align='center', label='Accuracy', color='blue',edgecolor='black')
plt.bar([x + 0.1 for x in x_values], data2, width=0.1, align='center', label='Precision', color='green',edgecolor='black')
plt.bar([x + 0.2 for x in x_values], data3, width=0.1, align='center', label='F1 Score', color='yellow',edgecolor='black')
plt.bar([x + 0.3 for x in x_values], data4, width=0.1, align='center', label='Recall', color='red',edgecolor='black')
plt.xticks(x_values, df['Model-Code'])
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
plt.figure(figsize=(10,5))
plt.title("All Classification Metrics of Different Clinical NER Models")
plt.plot(df['Accuracy'],marker='o',markerfacecolor='black',color='blue',linestyle='--',label='Accuracy')
plt.plot(df['Precision'],marker='o',markerfacecolor='black',color='green',linestyle='--',label='Precision')
plt.plot(df['F1 Score'],marker='o',markerfacecolor='black',color='yellow',linestyle='--',label='F1 Score')
plt.plot(df['Recall'],marker='o',markerfacecolor='black',color='red',linestyle='--',label='Recall')
plt.xticks(x_values, df['Model-Code'])
plt.xlabel('Models ----->')
plt.ylabel('Values ----->')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```
